```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

data = pd.read_csv("./creditcard.csv")
data.head(5)
```

```python
count_classes = pd.value_counts(data['Class'], sort=True).sort_index()
count_classes.plot(kind='bar')
plt.title("Fraud class histogram")
plt.xlabel("class")
plt.ylabel("frequency")
```

```python
# normalication
from sklearn.preprocessing import StandardScaler
data['normAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
data = data.drop(['Time', 'Amount'], axis=1)
data.head()
```

```python
# under sample
x = data.loc[:, data.columns != 'Class']
y = data.loc[:, data.columns == 'Class']
# Number of data points in the minority class
number_record_fraud = len(data[data['Class']==1])
fraud_indices = np.array(data[data['Class'] == 1].index)

normal_indices = data[data['Class'] == 0].index

random_normal_indices = np.random.choice(normal_indices, number_record_fraud, replace=False)
random_normal_indices = np.array(random_normal_indices)

# Appending the 2 indices
under_sample_indices = np.concatenate([fraud_indices, random_normal_indices])

# Under sample datasets
under_sample_data = data.iloc[under_sample_indices, :]

x_under_sample = under_sample_data.loc[:, under_sample_data.columns != 'Class']
y_under_sample = under_sample_data.loc[:, under_sample_data.columns == 'Class']

print("Percentage of normal transaction: ", len(under_sample_data[under_sample_data['Class']==0]) / len(under_sample_data))
print("Percentage of fraud transaction: ", len(under_sample_data[under_sample_data['Class']==1]) / len(under_sample_data))
print("Total number of transaction: ", len(under_sample_data))
# x.head()
# y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train_whole_datasets, X_test_whole_datasets, y_train_whole_datasets, y_test_whole_datasets = \
    train_test_split(x, y, test_size=0.3, random_state=0)
print("Number of whole transaction train datasets: ", len(X_train_whole_datasets))
print("Number of whole transaction teset datasets: ", len(X_test_whole_datasets))
print("Total number of transaction: ", len(X_train_whole_datasets) + len(X_test_whole_datasets))

#Under sample dataset
X_train, X_test, y_train, y_test = train_test_split(x_under_sample, y_under_sample,test_size=0.3, random_state=0)

print("Number transaction train datasets: ", len(X_train))
print("Number transaction teset datasets: ", len(X_test))
print("Total number of transaction: ", len(X_train) + len(X_test))
```
